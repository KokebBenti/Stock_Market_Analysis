```python
#Get the data relevant to the company we want
Companies=["AAPL","AMZN","GOOG","FB","NVDA","TSLA"]
News={}
for d in range (0,len(Companies)):
  News[d]=Sentiment.pick_stock(df,Companies[d])
News[3]
```

```python
#Import Stock Data
Companies=["AAPL","AMZN","GOOG","MSFT","NVDA","TSLA"]
Stocks={}
for s in range (0,len(Companies)):
 Stocks[s]=Correlation.import_stocks(Companies[s])
```

```python
#Aligning sentiment scores with the stock closing prices on the same date
Stocks_co={}
for m in range (0,len(Companies)):
 Stocks_co[m]=News[m].join(Stocks[m]['Adj Close'])
```

```python
#Calculate Daily Stock Returns
for u in range (0,len(Companies)):
 Stocks_co[u]=Correlation.DR(Stocks_co[u])
```

```python
#Calculate correlation
correlation= Stocks_co[1]['sentiment'].corr(Stocks_co[1]['Daily Returns'])
print("Correlation between sentiment and daily stock returns:", correlation)
```

```python
correlation = Stocks_co[2]['sentiment'].corr(Stocks_co[2]['Adj Close'])
print("Correlation between sentiment and closing:", correlation)
```

```python
for r in range(0,len(Companies)):
 import seaborn as sns
 import matplotlib.pyplot as plt

 # Select only the numeric columns you want to compare
 corr_data = Stocks_co[r][['sentiment', 'Adj Close', 'Daily Returns']]

 # Calculate correlation matrix
 corr_matrix = corr_data.corr(method='pearson')  # You can also try 'spearman' or 'kendall'

 # Plot heatmap
 plt.figure(figsize=(8, 6))
 sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
 plt.title(f'{Companies[r]} Correlation Matrix')
 plt.show()
```
